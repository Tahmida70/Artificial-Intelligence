# Exercise 4: Check Palindrome Number

```python
num = int (input("Enter a number:"))
original = num
reverse = 0
while num>0:
  digit = num % 10
  reverse = reverse * 10 + digit
  num = num//10
  if original == reverse:
     print("Palindrome number")
  else:
      print("not a palindrome number")
```

### The range() Function
If you do need to iterate over a sequence of numbers, the built-in function range() comes in handy. It generates arithmetic progressions:

```python
# repeated for 0 to 4
# for(i=0;i<5;i++)
for i in range(5):
    print(i)
```

<p style="text-align: justify">The given end point is never part of the generated sequence; range(10) generates 10 values, the legal indices for items of a sequence of length 10. It is possible to let the range start at another number, or to specify a different increment (even negative; sometimes this is called the ‘step’):</p>

```python
list(range(5, 10))
```

```python
list(range(0, 10, 3))
```

```python
list(range(10, 2, -1))
```

```python
list(range(-10, -100, -30))
```

# Functions
<p style="text-align: justify">The keyword def introduces a function definition. It must be followed by the function name and the parenthesized list of formal parameters. The statements that form the body of the function start at the next line, and must be indented.</p>

```python
def operate(a, b):
  print(str(a) + ' + ' + str(b) + ' = ' + str(a+b))
  print(str(a) + ' - ' + str(b) + ' = ' + str(a-b))
  print(str(a) + ' x ' + str(b) + ' = ' + str(a*b))
  print(str(a) + ' / ' + str(b) + ' = ' + str(a/b))
```

```python
operate(91,5)
```

```python
operate(5,2)
```

```python
def fib(n):    # write Fibonacci series up to n
    """Print a Fibonacci series up to n."""
    a, b = 0, 1
    while a < n:
        print(a, end=' ')
        a, b = b, a+b
        #print()
```

```python
# Now call the function we just defined:
fib(50)
```

```python
# add list values
def addValues(num):
    sum = 0
    for n in num:
        sum = sum + n
    return sum
```

<b>Exercise 4:</b> Define a <b>method/function</b> which will take two lists of numbers and returns a list (sum of two lists)
<p>Example:<br>
    Input: [3,4,5,1] and [6,7,2,8]<br>
    output: [9,11,7,9]</p>


```python
def sum_lists(list1,list2):
  result =[]
  for i in range(len(list1)):
    result .append(list1[i]+list2[i])
    return result
    print(sum_lists([3,4,5,1],[6,7,2,8]))
```

<b> Exercise 5:</b> Write a <b>method</b> which will take the <b>full name</b> as input and return both the first name and last name.
<p>Example:<br>
    Input: Arafat Rahman Sunny<br>
    Output: First name: Arafat, Last name: Sunny</p>

```python
def get_name(full_name):
  names = full_name.split()
  first_name = names[0]
  last_name = names[-1]
  return first_name, last_name
  print(get_name('Kazi Tahmida'))
```
